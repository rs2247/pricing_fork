```python
order_query = """
with active_experiment_pefs as (
    select 
        experiment_pre_enrollment_fees.*
        ,fee_experiment_alternatives.fee_experiment_id
    from
        querobolsa_production.experiment_pre_enrollment_fees 
        left join querobolsa_production.fee_experiment_alternatives 
            on fee_experiment_alternatives.id = experiment_pre_enrollment_fees.fee_experiment_alternative_id
        join querobolsa_production.fee_experiments on fee_experiments.id = fee_experiment_alternatives.fee_experiment_id
    where
        fee_experiments.end_date > '2019-10-01'
), 
base_orders as (
    select distinct
        orders.id as order_id
        ,pre_enrollment_fees.id as pef_id
        ,orders.checkout_step
        ,orders.initiated_at
        ,offers.id as offer_id
        ,offers.course_id
        ,offers.enabled offers_enabled
        ,offers.visible offers_visible
        ,courses.canonical_course_id
--        ,coalesce(active_fee_experiment_offers.fee_experiment_id,-1) fee_experiment_id
        ,active_experiment_pefs.fee_experiment_id
        ,courses.university_id
        ,kinds.parent_id as course_kind
        ,levels.parent_id as course_level
        ,offers.offered_price
        ,university_offers.max_payments
        ,pre_enrollment_fees.value as pef
        ,pre_enrollment_fees.value
            /(offers.offered_price
            *(case when university_offers.max_payments<=1
              then 0.
            when university_offers.max_payments<=12
              then 9*(0.09884+0.0846312*(university_offers.max_payments-2)) 
            when university_offers.max_payments<=24
              then 9*(0.94515+0.0603743*(university_offers.max_payments-12)) 
            when university_offers.max_payments<=36
              then 9*(1.66965+0.0436079*(university_offers.max_payments-24)) 
            when university_offers.max_payments<=48
              then 9*(2.19294+0.0357802*(university_offers.max_payments-36)) 
            when university_offers.max_payments<=60
              then 9*(2.62230+0.0307322*(university_offers.max_payments-48))
            else 2.69199 end)) take_rate  
        ,case when orders.checkout_step='paid' and coupon_exchanges.id is null
              then orders.id else null end as paid_order_id
        ,case when orders.checkout_step='paid' and coupon_exchanges.id is null
            then pre_enrollment_fees.value else 0 end as revenue


    from 
        querobolsa_production.orders
        join querobolsa_production.order_origins on order_origins.order_id=orders.id
        join querobolsa_production.line_items on line_items.order_id=orders.id
        join querobolsa_production.offers on offers.id=line_items.offer_id
        join querobolsa_production.university_offers on university_offers.id=offers.university_offer_id
        join querobolsa_production.courses on courses.id=offers.course_id
        join querobolsa_production.pre_enrollment_fees on pre_enrollment_fees.id=line_items.pre_enrollment_fee_id
        left join active_experiment_pefs 
            on active_experiment_pefs.pre_enrollment_fee_id = pre_enrollment_fees.id 
        left join querobolsa_production.coupons on coupons.order_id=orders.id
        left join querobolsa_production.refund_operations on refund_operations.coupon_id=coupons.id
        left join querobolsa_production.coupon_exchanges on coupon_exchanges.to_coupon_id=coupons.id
        left join (select * from querobolsa_production.kinds where parent_id is not null) kinds on kinds.name = courses.kind
        left join (select * from querobolsa_production.levels where parent_id is not null) levels on levels.name = courses.level

    where
        1=1
    --    and offers.enabled
    --    and offers.visible
          and offers.billing_integration = False -- no QAP
    --    and university_offers.enrollment_semester in ('2020.1')
        and scholar_semester(orders.initiated_at) = '2020.1'
        and orders.checkout_step <> 'initiated'
)
select
    *
from
    base_orders
"""
aux = spark.sql(order_query).toPandas()
aux.head()
```

```python
def expected_months(max_payments):
    if max_payments<=1:
        return 0.
    elif max_payments<=12:
        return 9*(0.09884+0.0846312*(max_payments-2)) 
    elif max_payments<=24:
        return 9*(0.94515+0.0603743*(max_payments-12))
    elif max_payments<=36:
        return 9*(1.66965+0.0436079*(max_payments-24)) 
    elif max_payments<=48:
        return 9*(2.19294+0.0357802*(max_payments-36)) 
    elif max_payments<=60:
        return 9*(2.62230+0.0307322*(max_payments-48))
    else:
        return 2.69199 
    

def pef2take_rate(df,price_cols):
    for price_col in price_cols:
        df[price_col + '_tr'] = df.apply(lambda x: x[price_col]/(x['avg_offered_price']*expected_months(x['max_payments'])),axis=1) 
    
    return df
```

```python
def limited_discount(x,discount):

    if x['course_kind']==1:
        limit = 50
    else:
        limit = 30
    
    return x['avg_pef']*(1-discount) if x['avg_pef']*discount < limit else (x['avg_pef']-limit)
```

```python
revenue_sim = df_sim.groupby(['course_kind','course_level','perc_gain_orders'])[[price_col + '_revenue' for price_col in list_remove(sim_prices,'avg_pef')] + ['baseline_revenue']].agg('sum').reset_index()
avg_pefs = df_sim.groupby(['course_kind','course_level'])['avg_pef'].agg('mean').reset_index()
revenue_sim = pd.merge(avg_pefs,revenue_sim,on=['course_kind','course_level']).sort_values('baseline_revenue',ascending=False)
for price_col in list_remove(sim_prices,'avg_pef'):
    if price_col !='baseline_revenue':
        revenue_sim[price_col + '_delta_revenue'] = 100*revenue_sim[price_col + '_revenue']/revenue_sim['baseline_revenue'] -100

# revenue_sim =  revenue_sim.sort_values(['baseline_revenue','perc_gain_orders'],ascending=[False,True])
revenue_sim =  revenue_sim.sort_values(['course_level','course_kind','perc_gain_orders'])

revenue_sim_gains = revenue_sim.drop([price_col + '_revenue' for price_col in list_remove(sim_prices,'avg_pef')] + ['baseline_revenue'],axis=1)
revenue_sim_gains
```
