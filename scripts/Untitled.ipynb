```python

import matplotlib.pyplot as plt
def gain(pef,discount):
    p2 = pef*(1-discount)
    return (0.404 -1.68*p2 +3.86*p2*p2)/(0.404 -1.68*pef +3.86*pef*pef)

fig,ax= plt.subplots()
for PEF in [0.2,0.15,0.1]:
    discount = [.1*x for x in range(1,10)]

    gains = [gain(PEF,disc) for disc in discount]


    ax.plot(discount,gains,label = PEF )
    
    ax.set_ylabel('Ganho na conversão')
    ax.set_xlabel('Desconto no PEF')
ax.legend()
ax.plot([0.5,0.5],[1,1.7],color='k',linestyle='--')
```
